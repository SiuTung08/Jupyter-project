```python
def declined_deposits(freeze_period, c_fail_freeze):
    result = []
    result.clear()
    freeze_start_time = datetime(year = 1, month = 1, day = 1, hour = 0, minute = 0, second = 0)
    freeze_end_time = datetime(year = 1, month = 1, day = 1, hour = 0, minute = 0, second = 0)
    for i in df2["ID"]:
        if (i != 1): 
            #determine the freeze period
            if (df2.at[i-1,"consecutive_fails"] == c_fail_freeze):
                freeze_start_time = df2.at[i-1,"txn_timestamp"]
                freeze_end_time = df2.at[i-1,"txn_timestamp"] + pd.DateOffset(hours=freeze_period)
            if (df2.at[i-2,"player_id"]==df2.at[i-1,"player_id"]):
                if (df2.at[i-1,"txn_timestamp"] > freeze_start_time and df2.at[i-1,"txn_timestamp"] <= freeze_end_time): 
                    df2.at[i-1,'freezed'] = True
                else:
                    df2.at[i-1,'freezed'] = False

            if (df2.at[i-2,"player_id"] != df2.at[i-1,"player_id"]):
                freeze_start_time = datetime(year = 1, month = 1, day = 1, hour = 0, minute = 0, second = 0)
                freeze_end_time = datetime(year = 1, month = 1, day = 1, hour = 0, minute = 0, second = 0)
                df2.at[i-1,'freezed'] = False
        else:
            if (df2.at[i-1,"consecutive_fails"] == c_fail_freeze):
                freeze_start_time = df2.at[i-1,"txn_timestamp"]
                freeze_end_time = df2.at[i-1,"txn_timestamp"] + pd.DateOffset(hours=freeze_period)
            df2.at[i-1,'freezed'] = False
    #calculate total number of impacted customer
    impacted_customers = []
    impacted_customers.clear()
    number_of_impacted_customers = 0

    for i in df2["ID"]:
        if (df2.at[i-1,'freezed'] == True and df2.deposit_status[i-1] != "Fail"):
            impacted_customers.append(df2.player_id[i-1])  

    number_of_impacted_customers = pd.Series(impacted_customers).nunique()
    #number_of_impacted_customers
    #calculate total number of impacted deposit    
    impacted_deposits = []
    impacted_deposits.clear()
    sum_of_impacted_deposits = 0

    for i in df2["ID"]:
        if (df2.at[i-1,'freezed'] == True and df2.deposit_status[i-1] != "Fail"):
            impacted_deposits.append(df2.attempted_deposit_amt_gbp[i-1])

    sum_of_impacted_deposits = sum(impacted_deposits)
    #sum_of_impacted_deposits
    
    result.append(number_of_impacted_customers)
    result.append(sum_of_impacted_deposits)
    
    return result
```
